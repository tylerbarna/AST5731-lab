```python
## importing modules
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import astropy
import astropy.units as u

from astropy.io import fits
```

```python
fitFile = fits.open('./data/GCs_NGC3379.fits')
datFile = pd.read_csv('./data/GCs_NGC3379.dat',sep='\s+')
csvFile = pd.read_csv('./data/GCs_NGC3379.csv')
```

```python
fitFile.info()
display(datFile.head()) 
display(csvFile.head()) 
```

```python
## plot the Rmag vs V-R color  
plt.figure(figsize=(10,5))
plt.scatter(csvFile['Rmag'],csvFile['V-R'],s=10)
plt.xlabel('Rmag')
plt.ylabel('V-R')
plt.title('Rmag vs V-R')
```

```python
## plot the Ypos vs Xpos from csvFile
plt.figure(figsize=(10,5))
plt.scatter(csvFile['Xpos'],csvFile['Ypos'],
c=csvFile['V-R'], cmap=plt.cm.CMRmap, s=10)
## add colorbar to the plot
plt.colorbar(label='V-R',orientation='vertical')
plt.xlabel('Xpos')
plt.ylabel('Ypos')
plt.title('Xpos vs Ypos')
```

```python
## plot a normalized histogram of Rmag
plt.figure(figsize=(10,5))
plt.hist(csvFile['Rmag'], density=True, range=[16,23]);
## using automatic binning in histogram for more stark comparison between the two methods
plt.title('Matplotlib Histogram of Rmag')
plt.show();
## plot a normalized histogram of Rmag with seaborn.kdeplot
plt.figure(figsize=(10,5))
sns.kdeplot(csvFile['Rmag'], shade=True, clip=[16,23])
plt.title('Seaborn KDEplot of Rmag')
plt.show();
```

```python
## plot a normalized histogram of V-R with maptlotlib
plt.figure(figsize=(10,5))
plt.hist(csvFile['V-R'], density=True,range=[0,1.0]);
plt.xlabel('V-R')
plt.ylabel('Frequency')
plt.title('Matplotlib Histogram of V-R')
plt.show();
## plot a normalized histogram of V-R with seaborn.kdeplot
plt.figure(figsize=(10,5))
sns.kdeplot(csvFile['V-R'], shade=True,clip=(0,1.0));
plt.title('Seaborn KDEplot of V-R')
plt.show();
```

```python
## Realized I was doing 1D histograms, so I made 2D histograms below but left the 1D histograms above

## 2D histogram of Xpos and Ypos with matplotlib
plt.figure(figsize=(5,5))
plt.hist2d(csvFile['Xpos'],csvFile['Ypos'],range=((-100,100),(-100,100)));
plt.title('Matplotlib 2D Histogram of Xpos and Ypos')
plt.xlabel('Xpos')
plt.ylabel('Ypos')
plt.show();

## 2D histogram of Xpos and Ypos with seaborn
plt.figure(figsize=(5,5))
sns.kdeplot(x=csvFile['Xpos'],y=csvFile['Ypos'],shade=True,clip=[(-100,100),(-100,100)]);
plt.title('Seaborn 2D KDEplot of Xpos and Ypos')
plt.xlabel('Xpos')
plt.ylabel('Ypos')
plt.show();
```

```python
## 2D histogram of Rmag and V-R with matplotlib
plt.figure(figsize=(5,5))
plt.hist2d(csvFile['Rmag'],csvFile['V-R'], range=((16,23),(0,1.0)));
plt.title('Matplotlib 2D Histogram of Rmag and V-R');
plt.xlabel('Rmag');
plt.ylabel('V-R');
plt.show();

## 2D histogram of Rmag and V-R with seaborn
plt.figure(figsize=(5,5))
sns.kdeplot(x=csvFile['Rmag'],y=csvFile['V-R'],shade=True,clip=[(16,23),(0,1.0)]);
plt.title('Seaborn 2D Histogram of Rmag and V-R');
plt.xlabel('Rmag');
plt.ylabel('V-R');
plt.show();
```
